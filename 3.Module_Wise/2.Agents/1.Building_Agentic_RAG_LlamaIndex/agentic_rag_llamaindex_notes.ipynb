### L3 - Building Agent Reasoning Loop

Till now we have done "Given a query , Call right tool with right parameters and get back the response. But this is still quit limiting . What if user ask complex question consisting of multiple steps or vague question that needs clarification.

In this lesson you will learn how to define a complete Agent Reasoning Loop. Instead of tool calling in single shot setting.

An agent is able to reason over tools and multiple steps . We will use Function calling Agent Implementation.Which is an agent that natively integrates with function calling capabilities of LLM 


Function calling Agent 

In LlamaIndex -- Agent consist of two main components 
- Agent Worker - Executing next step for a given agent
- Agent Runner - Overall task dispatcher. Which is responsible for creating task . Orchestrating runs of agent worker on top of given task. Return back the final response to user.



Function calling agent workers primary responsibility - As given existing conversation history.Use funciton calling to decide the next quote or call. Call that tool and decide whether or not to return  a final response.
