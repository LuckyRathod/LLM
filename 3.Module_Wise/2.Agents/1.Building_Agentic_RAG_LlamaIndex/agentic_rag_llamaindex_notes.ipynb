## Building Agentic RAG with LlamaIndex 

Framework to build research agents capable of events to use reasoning and decision making over your data 

For Example :

Lets say you have set of research papers on some topic . And you want to pull out the parts relevant to a question you ask and you get synthesis of what papers say . This is a complex request and it requires multiple steps of processing .

As we know that standard RAG pipeline is good for simple questions over small set of documents and works by retrieving some context. Sticking that context to llm prompt and then just calling LLM single time .

We will create Autonomous Research agent. We will learn following things :
1. Routing - Decision making to route requests to multiple tools 
2. Tools use - Interface for agents to selected tool . And also generate right arguments for that tool
3. Multistep reasoning with Range of Tools using LLM

#### Lesson 1 - Router Engine 

Given a Query - Router will pick one of several query entrants to execute a query . We will build a simple router over a single document that can handle both question answering and summarization.



We have 2 Indices . Each indices will have different retrieval methods . 

Vector Index



Summary Index 



There are three important points before defining router  :

1. Indexes : Nodes will be stored in sperate indexes
2. Query Engine : Each Query Engine represents overall query interface over the data thats stored in index .Each query engine is good for certain type of questions which is a very good use case of Router . Which can Route dynamically between this Query Engines
3. Query Tool : Its just a query engine with metadata . Its a description of what types of question the tool can answer.
